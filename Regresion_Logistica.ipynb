```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

data = pd.read_csv("C:/Users/User/Desktop/framingham.csv")

data
```

```python
data[['BMI','TenYearCHD']].plot.scatter(x='BMI',y='TenYearCHD')
```

```python
w = 0.09
b = -3.6
```

```python
x = np.linspace(0,data['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
data.plot.scatter(x='BMI',y='TenYearCHD')
plt.plot(x, y, color='black')
plt.ylim(0,data['TenYearCHD'].max()*1.1)
plt.scatter(x, y, color='#A9E2F3')
# plt.grid()
plt.xlabel('Riesgo cardiovascular')
plt.ylabel('Riesgo cardiovascular 1:Positivo 0:Negativo')
plt.show()
```
