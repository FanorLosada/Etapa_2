```python
import matplotlib.pyplot as plt
from sklearn import linear_model # usando sklear para saber los valores optimos
import seaborn as sns
import numpy as np
import pandas as pd
 
datos = pd.read_csv("C:/Users/User/Desktop/ingreso.csv")
datos
```

```python
plt.ylabel("Dinero")
plt.xlabel("Promedio horas trabajadas semanales")
plt.scatter(datos["horas"],datos["ingreso"],color="purple")
plt.show()
```

```python
regression = linear_model.LinearRegression()

hours = datos["horas"].values.reshape((-1, 1))

model = regression.fit(hours, datos["ingreso"])

print("IntersecciÃ³n (b)", model.intercept_)
print("Pendiente (m)", model.coef_)

entrance = [[39], [40], [43]]

model.predict(entrance)

plt.scatter(entrance, model.predict(entrance), color="blue")
plt.plot(entrance, model.predict(entrance), color="black")

plt.ylabel("Dinero")
plt.xlabel("Promedio horas trabajadas semanales")
plt.scatter(datos["horas"], datos["ingreso"], color="purple", alpha=0.55)
plt.show()
```
